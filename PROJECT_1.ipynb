```python
model=genai.GenerativeModel("models/gemini-2.5-flash")
```

```python
chatbot_name = "Chatbot"

while True:
    user_input = input("You: ")
    processed_input = user_input.strip().lower()

    if processed_input == "exit":
        print(f"{chatbot_name}: Goodbye!")
        break
    elif processed_input == "rename":
        new_name = input("Enter new chatbot name: ")
        chatbot_name = new_name.capitalize() #Captilize the first letter
        print(f"Chatbot name updated to {chatbot_name}.")
        continue

    response = model.generate_content(user_input, stream=True)

    full_response_text = ""
    print(f"{chatbot_name}: ", end="")
    for chunk in response:
        if chunk.text:
            print(chunk.text, end="")
            full_response_text += chunk.text
    print() # Add a newline at the end 

    if not full_response_text.strip():
        print(f"{chatbot_name}: I do not understand")
```
